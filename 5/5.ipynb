# Lesson 5

````python
import numpy as np
from matplotlib import pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
````

````python
def error(ave, ave2, n):
    if n == 0:
        return 0
    return np.sqrt((ave2 - ave*ave) / n)

def blocks_value(data):
    total = 0.
    total2 = 0.
    y = np.zeros((len(data), 2))
    for i in range(len(data)):
        total = total + data[i]
        total2 = total2 + data[i]*data[i]
        ave = total / (i+1)
        ave2 = total2 / (i+1)
        y[i, 0] = ave
        y[i, 1] = error(ave, ave2, i)
    return y
````

## Exercise 5.1

In this exercise, we use the Metropolis algorithm to sample from the probability distributions corresponding to two states of a hydrogen atom: $|\psi_{100}|^2$ and $|\psi_{210}|^2$. The expected values are $<r_{100}>=1.5a_0$ and $<r_{210}>=5a_0$, where $a_0$ is the Bohr radius.

````python
with open("data/5.1-info.dat") as file:
    N_blocks = int(file.readline())
    N_throws_per_block = int(file.readline())
    accept_rate_psi100_unif = file.readline()
    accept_rate_psi100_gauss = file.readline()
    accept_rate_psi210_unif = file.readline()
    accept_rate_psi210_gauss = file.readline()
````

## $\psi_{100}$

### With uniform Metropolis

````python
ave = np.loadtxt("data/5.1-ave-psi100-unif.dat")

x = np.arange(N_blocks) * N_throws_per_block
y = blocks_value(ave)
plt.errorbar(x, y[:,0]-3/2, yerr=y[:,1])
plt.xlabel(r"#throws")
plt.ylabel(r"$<r/a_0>-3/2$")
plt.grid(True)
plt.show()
print(f"<r/a_0>-1 = {y[-1, 0]-3/2:.2f} ± {y[-1, 1]:.2f}")
print(f"Metropolis acceptance rate = {accept_rate_psi100_unif}")
````

````python
data = np.loadtxt("data/5.1-psi100-unif.dat")

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:,0], data[:,1], data[:,2], c=data[:,2], marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
````
