```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
cities = np.loadtxt("data/9.1-circ-cities.dat")
out = np.loadtxt("data/9.1-circ-out.dat")
best = np.loadtxt("data/9.1-circ-best.dat", dtype="int")

x = np.arange(out[:, 0].size)
plt.plot(x, out[:, 0], label="Best");
plt.plot(x, out[:, 1], label="Average of best half");
plt.legend()
plt.show()
plt.plot(cities[best, 1], cities[best, 2]);
plt.show()
```
