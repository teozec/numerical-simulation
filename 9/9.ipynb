````python
import numpy as np
from matplotlib import pyplot as plt
````

## Exercise 9.1

In this exercise, we use a Genetic Algorithm to solve the Travelling Salesman Problem for cities placed on a circumference and inside a square.

### Circumference

````python
cities = np.loadtxt("data/9.1-circ-cities.dat")
out = np.loadtxt("data/9.1-circ-out.dat")
best = np.loadtxt("data/9.1-circ-best.dat", dtype="int")
best = np.append(best, best[0])
````

````python
x = np.arange(out[:, 0].size)
plt.plot(x, out[:, 0], label="Best");
plt.plot(x, out[:, 1], label="Average of best half");
plt.legend()

plt.grid()
plt.title("Path cost ($L_1$)")
plt.xlabel("Generation")
plt.ylabel("Cost")

plt.show()
````

````python
plt.figure(figsize=(6, 6))
plt.plot(cities[best, 1], cities[best, 2]);
plt.grid()
plt.title("Path")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
````
